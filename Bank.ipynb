```python
X = data.iloc[:,[0,1,2,3,4,5,6,7,8,11,12,13,14,15]].values
y = data.iloc[:,16].values
```

```python
X.shape, y.shape
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
data.columns
```

```python
job = LabelEncoder() #1
marital = LabelEncoder() # 2
education = LabelEncoder() # 3
default = LabelEncoder()  # 4
housing = LabelEncoder() # 6
loan = LabelEncoder() # 7
contact = LabelEncoder() # 8
poutcome = LabelEncoder() # 13
```

```python
X[:, 1] = job.fit_transform(X[:, 1])
X[:, 2] = marital.fit_transform(X[:, 2])
X[:, 3] = education.fit_transform(X[:, 3])
X[:, 4] = default.fit_transform(X[:, 4])
X[:, 6] = housing.fit_transform(X[:, 6])
X[:, 7] = loan.fit_transform(X[:, 7])
X[:, 8] = contact.fit_transform(X[:, 8])
X[:, 13] = poutcome.fit_transform(X[:, 13])
```

```python
X[:10]
```

```python
print("jobs: ", job.classes_)
print("marital: ", marital.classes_)
print("education: ", education.classes_)
print("default: ", default.classes_)
print("housing: ", housing.classes_)
print("loan: ", loan.classes_)
print("contact: ", contact.classes_)
print("poutcome", poutcome.classes_)
```

```python
one_hot_encoder = OneHotEncoder(categorical_features=[1,2,3,4,6,7,8,13])
X = one_hot_encoder.fit_transform(X).toarray()
```

```python
X.shape
```

```python
X = np.delete(X, [0,12,15,19,21,23,25,28], axis= 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=500)
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Dropout
from keras.optimizers import Adam
```

```python
y_label = LabelEncoder()
y = y_label.fit_transform(y)
y
```

```python
model = Sequential()
model.add(Dense(64, input_dim= 30, activation="sigmoid"))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(64, activation="sigmoid"))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(64, activation="sigmoid"))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(32, activation="sigmoid"))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(32, activation="sigmoid"))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(8, activation="sigmoid"))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(1, activation="sigmoid"))
model.compile(optimizer=Adam(lr=0.00001), metrics=["accuracy"], loss= "binary_crossentropy")
```

```python
model.fit(X,y, validation_split=0.4, epochs=40, verbose=0);
```

```python
plt.plot(model.history.history["acc"]);
plt.plot(model.history.history["val_acc"]);
```
